```
import pandas as pd
import matplotlib.pyplot as plt
```

# Plot daily visitors 2019 vs 2020

```
visit_2019_df = pd.read_csv('https://raw.githubusercontent.com/kywch/geo-colab/master/process/daily_visits_2019.csv')
visit_2019_df['date'] = pd.date_range(start='3/1/2019', end='8/31/2019')
visit_2019_df.set_index('date')
```

```
visit_2020_df = pd.read_csv('https://raw.githubusercontent.com/kywch/geo-colab/master/process/daily_visits_2020.csv')
visit_2020_df['date'] = pd.date_range(start='3/1/2020', end='8/31/2020')
visit_2020_df.set_index('date')
```

```
place_types = visit_2020_df.columns[1:15]
```

```
fig, axs = plt.subplots(5, 3, figsize=(17, 25), sharex='col')

for row in range(0,5):
  for col in range(0,3):
    idx = 3*row + col
    if idx < len(place_types):
      axs[row, col].plot(visit_2019_df[place_types[idx]], '-', label = '2019')
      axs[row, col].plot(visit_2020_df[place_types[idx]], '-', label = '2020')      
      axs[row, col].set_title(place_types[idx])
      axs[row, col].set_ylim(bottom = 0)


#  ax[ii].legend()
#  ax[ii].xlabel('Days since March 1st')
#  ax[ii].ylabel('Total number of visits')
#  ax[ii].title(place_types[ii])
```
